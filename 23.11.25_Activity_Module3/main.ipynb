Remove the columns with rows that has null values that are higher than 10%

```python
df_col_clean = df.dropna(axis=1, thresh=df.shape[0]*.9).copy()

# drop the upc column (it's not useful for our analysis)
# for context UPC stands for Universal Product Code and is a unique identifier for a product
df_col_clean.drop('upc', axis=1, inplace=True)

# drop the prices.currency column (as all values will be converted to PHP)
df_col_clean.drop('prices.currency', axis=1, inplace=True)

df_col_clean.isnull().sum()
```

Convert the shoe size into millimeters <i>to distinguish what is the smaller to largest</i>

```python

#https://en.wikipedia.org/wiki/Shoe_size

def uk_to_mm(size):
    """
    Converts a UK shoe size to mm.

    Args:
        size: The UK shoe size (float)

    Returns:
        The shoe size in mm (int)
    """
    return (25.4/3) * (size + 23)


def eur_to_mm(size):
    """
    Converts a EUR shoe size to mm.

    Args:
        size: The EUR shoe size (float)

    Returns:
        The shoe size in mm (int)
    """
    return (20/3) * (size - 2)

def us_to_mm(size):
    """
    Converts a US shoe size to mm.

    Args:
        size: The US shoe size (float)

    Returns:
        The shoe size in mm (int)
    """
    return uk_to_mm(size + 2)


for i, v in enumerate(df_col_clean['prices.size']):
    spl = str(v).upper().split('/')


    if spl[0] in ('S', 'SMALL'):
        df_col_clean.loc[i, 'prices.size'] = uk_to_mm(6)
        continue
    elif spl[0] in ('M', 'MEDIUM'):
        df_col_clean.loc[i, 'prices.size'] = uk_to_mm(8)
        continue
    elif spl[0] in ('L', 'LARGE'):
        df_col_clean.loc[i, 'prices.size'] = uk_to_mm(10)
        continue

    if len(spl) > 1:
        #split to dict
        spl = dict(s.split(' ') for s in spl)
        
        if 'US' in spl:
            df_col_clean.loc[i, 'prices.size'] = us_to_mm(float(spl['US']))
            continue
        if 'WOMENS' in spl:
            df_col_clean.loc[i, 'prices.size'] = us_to_mm(float(spl['WOMENS']))
            continue
    else:
        spl = v.split(' ')
        val = float(spl[0])

        if val > 30:
            val = eur_to_mm(float(spl[0]))
        else:
            val = us_to_mm(float(spl[0]))
            
        df_col_clean.loc[i, 'prices.size'] = val
```

Set the missing values for the <code>price.dateAdded</code> with the average<br/>Then convert the <i>dates</i> into <i>epoch</i>

```python
#convert date to epoch
date_features = ['dateAdded', 'dateUpdated', 'prices.dateAdded', 'prices.dateSeen']



for feature in date_features:
    df_col_clean[feature] = pd.to_datetime(
        df_col_clean[feature]
            .map(lambda x: x if pd.isnull(x) or not isinstance(x, str) else x.split(',')[0]))
    
    # fill the NaN values with the mean
    df_col_clean[feature] = df_col_clean[feature].fillna(df_col_clean[feature].mean())
    
    # if the value is not NaN, convert to epoch using lambda
    df_col_clean[feature] = df_col_clean[feature].map(lambda x: x if pd.isnull(x) else int(x.toordinal()))

```

Convert <code>prices.amountMax</code> and <code>prices.amountMin</code> into PHP

```python
# in order to convert the prices to PHP, we need to convert the currency first
# we will use the currencyconverter library to convert the currencies for more accuracy
%pip install currencyconverter
from currency_converter import CurrencyConverter
```

```python
# convert the prices.amountMax and prices.amountMin to PHP
USD_PHP = CurrencyConverter().convert(1, 'USD', 'PHP')

df_col_clean['prices.amountMax'] = df_col_clean['prices.amountMax'].map(lambda x: x * USD_PHP)
df_col_clean['prices.amountMin'] = df_col_clean['prices.amountMin'].map(lambda x: x * USD_PHP)
```

```python
df_col_clean['prices.amountMax'], df_col_clean['prices.amountMin']
```

Adding the average amount for the price

```python
# add column for price.amountAvg
df_col_clean['prices.amountAvg'] = (df_col_clean['prices.amountMax'] + df_col_clean['prices.amountMin']) / 2
```

ANOVA

```python
# create anova
import scipy.stats as stats
unqs = df_col_clean['prices.color'].unique()
```

```python
df_group_by_color = df_col_clean[['prices.color', 'prices.amountAvg']].groupby(['prices.color'])

lst = []

for color in unqs:
    lst.append(df_group_by_color.get_group(color)['prices.amountAvg'])

f_v, p_v = stats.f_oneway(*lst)

print("ANOVA RESULTS: f = ", f_v, "p = ", p_v)
```
