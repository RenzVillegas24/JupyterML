<h1>Creating a Dummy Variable for the Independent Variable</h1>
<h3>X for the Variables excluding <code>Salary</code></h3>

```python
from sklearn.preprocessing import OneHotEncoder

# get only the categorical columns in dfc_X
X_categorical = X.select_dtypes(include = ['object']) 

# print the categorical columns
print('Categorical columns:') 
display(X_categorical)

# get only the numerical columns in dfc_X
X_numerical = X.select_dtypes(include = ['int64', 'float64'])

# print the numerical columns
print('Numerical columns:')
display(X_numerical)

'''
    Other method to create dummy variables, different from the one used in the lecture, 
    but with the same result.
    We did not use ColumnTransformer in this case, 
    because it is not necessary for this method.
'''

# encode the categorical columns
enc = OneHotEncoder(
    sparse_output = False,  # returns a numpy array, not a sparse matrix
    drop='first',           # to avoid multicollinearity/dummy variable trap
    categories='auto')

# fit and transform the data
dfc_X_enc = pd.DataFrame(
    data    = enc.fit_transform(X_categorical),                 # return the encoded data
    columns = enc.get_feature_names_out(X_categorical.columns) # new column names
)

# display the encoded data
print('Encoded categorical columns:')
display(dfc_X_enc)

# combine the numerical and encoded categorical columns
dfc_X = pd.concat([X_numerical, dfc_X_enc], axis=1)

print('All columns after encoding:')
display(dfc_X)

#combine the dfc_norm and price
dfc_norm = pd.concat([dfc_X.copy(), Y], axis=1)

# save the encoded data to a csv file
dfc_norm.to_csv('Group 11_CompleteDataset.csv', index=False)
```

<h1>Backward Elimination Process for Linear Model Creation</h1>

```python
import statsmodels.regression.linear_model as lm

def feature_selection(Y, dfc_X):
    """
    Perform feature selection using the OLS (Ordinary Least Squares) method.
    
    Args:
        Y (array-like): The dependent variable.
        dfc_X (DataFrame): The independent variables.
    
    Returns:
        removed (list): The list of removed features.
        model (statsmodels.regression.linear_model.RegressionResultsWrapper): The final model.
    """
    # Fit the OLS model
    model = lm.OLS(endog=Y, exog=dfc_X).fit()

    # List of removed features
    removed = []

    while True:
        # Get the p-values for all features
        p_values = model.pvalues[1:]  # Exclude the intercept term

        # Find the feature with the highest p-value
        max_p_value = p_values.max()
        max_p_index = p_values.idxmax()

        # If the highest p-value is above the significance level (e.g., 0.05), 
        # remove the corresponding feature
        if max_p_value > 0.05:
            # add the removed feature to the list
            removed.append(max_p_index)
            # remove the feature from the data
            dfc_X = dfc_X.drop(max_p_index, axis=1)
            # fit the model again
            model = lm.OLS(endog=Y, exog=dfc_X).fit()
        else:
            break

    return removed, model

removed_features, final_model = feature_selection(Y, dfc_X)

# Print the removed features
print('Removed features:')
display(pd.DataFrame(removed_features, columns=['Index']))

# Print the final model summary
print('Final model summary:')
display(final_model.summary())
```

<h3>Conclusion</h3>
The program takes 756 trials until it successfully finds the best model.

```python
# print the lowest p-value and its index
print('Lowest p-value:', final_model.pvalues[1:].min(), '\nDummy variable:', final_model.pvalues[1:].idxmin())
```

```python
# print the R-squared value
print('R-squared:', final_model.rsquared)
```

```python
# print the Adjusted R-squared value
print('Adjusted R-squared:', final_model.rsquared_adj)
```
