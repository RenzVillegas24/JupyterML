<h1>Importing Required Libraries</h1>
This file contains all the libraries that are required for the project.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score, r2_score
import matplotlib.pyplot as plt


from pandas.plotting import scatter_matrix

# import different machine learning models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

# for showing the proper table
from IPython.display import display
```

```python
models = []
models.append(('LR', LogisticRegression(max_iter=1000000)))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
models.append(('NN', MLPClassifier()))
models
```

<h1>Loading and Viewing Dataset</h1>
Import the database for all the features.

```python
# read the data
df_norm = pd.read_csv('ECEN E3_Group 11_MP1 Part1_DW and EDA.csv')

print("Peeking into the data")
display(df_norm.head(10))

print("Statistical description of the data")
display(df_norm.describe())

'''
print("Histogram of the data")
df_norm.hist(figsize=(15,15))
display(plt.show())

print("Scatter matrix of the data")
scatter_matrix(df_norm)
display(plt.show())
'''

#
```

```python
# use label encoder to convert the categorical data to numerical data
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

# get the categorical columns
categorical_columns = df_norm.select_dtypes(include=['object']).columns
print("Categorical columns are: ") 
display(categorical_columns)

# apply label encoder to the categorical columns
df_norm[categorical_columns] = df_norm[categorical_columns].apply(lambda col: labelencoder.fit_transform(col.astype(str)))

print("Peeking into the data")
display(df_norm.head(10))
```

<h2>Independent and Dependent Variables.</h2>
Creates a matrix of independent variables and a matrix of dependent variables.

```python

# create a matrix of dependent variables
X_norm = df_norm.drop(['Price'], axis=1).values

print("Normal Independent Variable Array:")
display(pd.DataFrame(X_norm))

# create a matrix of independent variables
Y_norm = df_norm['Price'].values

# convert Y_norm using label encoder
Y_norm = labelencoder.fit_transform(Y_norm)

print("Normal Dependent Variable Array:")
display(pd.DataFrame(Y_norm))
```

<h2>Training and Testing Datasets</h2>
Split the data into training and testing datasets

```python
X_norm_train, X_norm_test, Y_norm_train, Y_norm_test = train_test_split(X_norm, Y_norm, test_size=0.2, random_state=0)

print("Normal X Train:")
display(pd.DataFrame(X_norm_train))

print("Normal X Test:")
display(pd.DataFrame(X_norm_test))

print("Normal Y Train:")
display(pd.DataFrame(Y_norm_train))

print("Normal Y Test:")
display(pd.DataFrame(Y_norm_test))
```

```python
# For Test Options and Evaluation Metric
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

results = []
names = []

for name, model in models:
	k_Fold = StratifiedKFold(n_splits=10, shuffle=True, random_state=None)
	cv_results = cross_val_score(model, X_norm, Y_norm, cv=k_Fold, scoring='accuracy')
	results.append(cv_results)
	names.append(name)
	printed_results = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
	print(printed_results)
```
